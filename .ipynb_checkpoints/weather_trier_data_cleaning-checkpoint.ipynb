# Weather in Trier Germany

Because my basement together with many others was flooded two times in the last two years due to heavy rainfall in my hometown Trier Germany, I wanted to investigate the weather in Trier and better understand it. This was a good opportunity to use some methods for time series analysis and better understand the difference between classical methods to deep learning. The compared methods will be ARIMA and LSTM.

The collected data can be found on [wetterkontor](https://www.wetterkontor.de/de/wetter/deutschland/monatswerte-station.asp).

In this notebook, we will just clean the data. A good tutorial for this procedure in a weather context be found for the [india rainfall](https://www.kaggle.com/biphili/india-rainfall-kerala-flood) on Kaggle and its responses.

## Importing moduls and data

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# drop warning for assigning slices of a dataframe
pd.options.mode.chained_assignment = None  # default='warn'

# Defaults
plt.rcParams['figure.figsize'] = (55, 10.0)
plt.rcParams.update({'font.size': 12})
plt.style.use('seaborn')
```

```python
print(os.listdir("data"))
weather_raw = pd.read_excel('data/weather_trier.xlsx')
weather_raw.head()
```

## Cleaning the data and change column names

1. Only each third row contains useful data, so we drop the rest. 
2. Rename the columns to english names.
3. Change the data types

```python
weather = weather_raw[[k % 3 == 0 for k in range(len(weather_raw))]]

col_names = ['time', 'temperature', 'rain', 'sunshine']
weather.columns = col_names

weather.time = pd.to_datetime(weather.time)
for col in weather.columns[1:]:
    weather[col] = pd.to_numeric(weather[col], errors = 'coerce')
weather.head()
```

This looks way better, now we add some context. The column 'temperature' contains for each time the montly average temperature measured in degree Celsius in Trier Germany. The column 'rain' the monthly cumulated rainfall measured in mm and the column 'sunshine' the monthly cumulated sunshine hours. Thus we see that the most recent observation is not complete and we drop it. We change the order of the observations to have it for futher investigations timely uprising and reset the index.

```python
weather = weather[1:]
weather = weather[::-1]
weather.index = range(len(weather.index))
weather
```

## Summarize the dataset
We find missing values.

```python
weather.info()
```

```python
print('Rows    :', weather.shape[0])
print('Columns :', weather.shape[1])
print('\nFeatures:', weather.columns.tolist())
print('\nMissing values:',weather.isnull().values.sum())
print(weather.isnull().sum())
print('\nUnique values:',weather.nunique().sum())
print( weather.nunique())
```

## Handling the missing values
We fill the missing values with the monthly mean.

```python
import datetime as dt # for datetime manipulation

for k in range(1,13):
    month_bool = (weather.time.dt.month == k)
    weather[month_bool]= weather[month_bool].fillna(weather[month_bool].mean())
```

## General analysis

We are going to focus only temperature and rainfall, because we are most interested in these features. The methods can easily be adaped for other data and columns. First we plot all the data, which does not give us much insight.

```python
fig, ax = plt.subplots(2, 1)
fig.tight_layout(pad=3.0)
fig.suptitle('Weather in Trier Germany monthly from 1989 to 2020')

plt.subplot(2,1,1)
plt.plot('time', 'temperature', data = weather, label = 'actual temperature')
plt.plot(weather.time, weather.temperature.rolling(10).mean(), color = 'r', label = 'rolling mean order 10')
plt.title('Monthly average temperature (in degree Celsius)')
plt.ylim((-5,32))
plt.legend()

plt.subplot(2,1,2)
plt.plot('time', 'rain', data = weather, label = 'actual rainfall')
plt.plot(weather.time, weather.rain.rolling(10).mean(), color = 'r', label = 'rolling mean order 10')
plt.title('Monthly cumulated rainfall (in mm)')
plt.legend()

plt.show()
```

To get more insight we group by each year.

```python
fig, ax = plt.subplots(2, 1)
fig.tight_layout(pad=3.0)
fig.suptitle('Weather in Trier Germany yearly from 1989 to 2019')

plt.subplot(2,1,1)
weather.groupby(weather.time.dt.year).temperature.mean()[:-1].plot(marker = 'o', label = 'actual temperature')
weather.groupby(weather.time.dt.year).temperature.mean().rolling(8).mean()[:-1].plot(marker = 'o', color = 'r', label = 'rolling mean order 10')
plt.xlabel('')
plt.title('Yearly average temperature (in degree Celsius)')
plt.legend()

plt.subplot(2,1,2)
weather.groupby(weather.time.dt.year).rain.sum()[:-1].plot(marker = 'o', label = 'actual rainfall')
weather.groupby(weather.time.dt.year).rain.sum().rolling(8).mean()[:-1].plot(marker = 'o', color = 'r', label = 'rolling mean order 10')
plt.xlabel('')
plt.title('Yearly cumulated rainfall (in mm)')
plt.legend()

plt.show()
```

By the rolling means we now can see an increase in temperature and a decrease in rainfall, which both can be explained by global warming. 

## Categorized in seasons

Next we take a closer look at each season seperatly.

```python
seasons = {'winter': [12, 1, 2],
           'spring': [3, 4, 5],
           'summer': [6, 7, 8],
           'autumn': [9, 10, 11]}
cols = ['temperature', 'rain']

fig, axes = plt.subplots(4,2,sharex = True, figsize = (12,8))
fig.suptitle('Weather in Trier Germany by seasons from 1989 to 2019')


for i, key in zip(range(4), seasons.keys()):
    val = seasons[key]
    x_temp = (weather.groupby(weather.time.dt.year)
              .apply(lambda x: x.loc[x.time.dt.month.isin(val), 'temperature'].mean())[:-1])
    x_rain = (weather.groupby(weather.time.dt.year)
              .apply(lambda x: x.loc[x.time.dt.month.isin(val), 'rain'].sum())[:-1])
    ax1 = axes[i, 0]
    ax2 = axes[i, 1]
    ax1.plot(x_temp, marker = 'o', label = 'actual temperature')
    ax1.plot(x_temp.rolling(10).mean(),  color = 'red', label = 'rolling mean order 10')
    ax1.set_title('Average ' + key + ' temperature (in degree Celsius)')
    ax1.set_ylim((-3,25))
    ax1.legend()
    
    ax2.plot(x_rain, label = 'actual rainfall', marker = 'o')
    ax2.plot(x_rain.rolling(10).mean(),  color = 'red', label = 'rolling mean order 10')
    ax2.set_title('Cumulated '+ key +' rainfall (in mm)')
    ax2.set_ylim((0,350))
    ax2.legend()
```

```python
weather['month'] = weather.time.dt.month

plt.figure(figsize =  (15,5))
plt.suptitle("Monthly weather in Trier Germany ")
plt.subplot(1,3,1)
plt.title('Monthly average temperature')
sns.boxplot('month', 'temperature', data = weather)
plt.subplot(1,3,2)
plt.title('Monthly cumulative rainfall')
sns.boxplot('month', 'rain', data = weather)
plt.subplot(1,3,3)
plt.title('Monthly cumulative sunshine duration')
sns.boxplot('month', 'sunshine', data = weather)

plt.show()
```

# First time series analysis

The easiest prediction of the future value is the current value. We try each of the 12 predecessors as prediction and measure the error. All our other further prediction methods should be more accurate as this very basic one.

```python
from sklearn.metrics import mean_squared_error # for error calculation

def error_rain_shift(t = 1):
    rain_shift = weather.rain.shift(t)[t:]
    rain = weather.rain[t:]
    err = mean_squared_error(rain, rain_shift)
    return np.sqrt(err)
    

shifts = range(1,14)
errors = [error_rain_shift(t) for t in shifts]
plt.title('Prediction by predecessorts: order vs squareroot of mse')
plt.plot(shifts, errors, marker = 'o')
plt.xticks(shifts)
plt.xlabel('order')
plt.ylabel('squareroot of mse')
plt.show()
```

### Visualization for year 2019

```python
#import mathplotlib.mdates

start19 = weather[weather.time =='2019-01-01'].index[0]
end19 = start19 + 11
y2019 = weather.time[start19:end19]
rain19 = weather.rain[start19:end19]
plt.plot(y2019.dt.month, rain19, marker = 'o', label = 'actual rainfall')
for i in [1,5,10]:
    rain_i =  weather.rain[(start19 - i):(end19-i)]
    plt.plot(y2019.dt.month, rain_i, marker = 'o', 
             label = 'shift {}: sqrt(mse) = {:.2f}'.format(i, np.sqrt(mean_squared_error(rain19, rain_i))))

plt.legend()
plt.title('Comparison for 2019')
plt.xticks(y2019.dt.month)
plt.xlabel('month')
plt.show()
```
